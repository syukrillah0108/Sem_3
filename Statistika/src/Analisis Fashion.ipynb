```python
# Fungsi untuk menambahkan garis regresi linear ke plot
def plot_linear_regression(x, y, title):
    model = LinearRegression()
    model.fit(x.values.reshape(-1, 1), y)
    x_range = np.linspace(min(x), max(x), 100)
    y_pred = model.predict(x_range.reshape(-1, 1))
    plt.plot(x_range, y_pred, color='red', linestyle='--', linewidth=2)
    plt.title(title)
    plt.xlabel(x.name)
    plt.ylabel(y.name)
    plt.show()

# Pilih variabel x dan y untuk regresi linear
x_vars = ['Seberapa besar pengaruh merek fashion lokal dalam mempengaruhi keputusan anda untuk membeli produk mereka?',
          'Usia', 'Apakah harga produk fashion brand lokal mempengaruhi ?', 'Sejauh mana anda percaya bahwa merek fashion lokal menyediakan produk berkualitas?']
y_var = 'Apakah harga produk fashion brand lokal mempengaruhi ?'

try:
    data['Usia'] = data['Usia'].str.split('-').apply(lambda x: np.mean([int(i) for i in x])).astype(float)
except: None
```

```python
for x_var in x_vars:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x_var, y=y_var, hue='Gender', data=data, s=100)
    plot_linear_regression(data[x_var], data[y_var], f'Kesimpulan: \n{x_var} \ndan {y_var} berdasarkan Gender')
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk menambahkan garis regresi linear ke plot
def plot_linear_regression(x, y, title, x_label, y_label):
    model = LinearRegression()
    model.fit(x.values.reshape(-1, 1), y)
    x_range = np.linspace(min(x), max(x), 100)
    y_pred = model.predict(x_range.reshape(-1, 1))
    plt.plot(x_range, y_pred, color='red', linestyle='--', linewidth=2)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

# Pilih variabel x dan y untuk regresi linear
x_var = 'Seberapa sering anda membeli produk fashion lokal jika sedang mencari baju baru?'

# Plot regresi linear dengan orientasi x horizontal dan y vertikal
plt.figure(figsize=(12, 6))
sns.scatterplot(x=x_var, y='Apakah harga produk fashion brand lokal mempengaruhi ?', hue='Gender', data=data, s=100)
plot_linear_regression(data[x_var], data['Apakah harga produk fashion brand lokal mempengaruhi ?'],
                       f'Regresi Linear:\n {x_var} dan\n Apakah harga produk fashion brand lokal mempengaruhi ? (Orientasi X Horizontal)',
                       x_var,
                       'Apakah harga produk fashion brand lokal mempengaruhi ?')

# Plot regresi linear dengan orientasi x vertikal dan y horizontal
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Apakah harga produk fashion brand lokal mempengaruhi ?', y=x_var, hue='Gender', data=data, s=100)
plot_linear_regression(data['Apakah harga produk fashion brand lokal mempengaruhi ?'], data[x_var],
                       f'Regresi Linear: \nApakah harga produk fashion brand lokal mempengaruhi ? dan\n {x_var} (Orientasi Y Horizontal)',
                       'Apakah harga produk fashion brand lokal mempengaruhi ?',
                       x_var)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Fungsi untuk menambahkan garis regresi linear ke plot
def plot_linear_regression(x, y, title, orientation):
    model = LinearRegression()
    if orientation == 'horizontal':
        model.fit(y.values.reshape(-1, 1), x)
    else:
        model.fit(x.values.reshape(-1, 1), y)
    x_range = np.linspace(min(x), max(x), 100)
    y_pred = model.predict(x_range.reshape(-1, 1))

    plt.figure(figsize=(8, 6))
    if orientation == 'horizontal':
        plt.scatter(y, x, s=100)
        plt.plot(y_pred, x_range, color='red', linestyle='--', linewidth=2)
        plt.title(title)
        plt.xlabel("Seberapa sering anda membeli produk fashion \nlokal jika sedang mencari baju baru?")
        plt.ylabel("Seberapa besar pengaruh merek fashion lokal dalam \nmempengaruhi keputusan anda untuk membeli produk mereka?")
    else:
        plt.scatter(x, y, s=100)
        plt.plot(x_range, y_pred, color='red', linestyle='--', linewidth=2)
        plt.title(title)
        plt.xlabel("Seberapa besar pengaruh merek fashion lokal dalam \nmempengaruhi keputusan anda untuk membeli produk mereka?")
        plt.ylabel("Seberapa sering anda membeli produk fashion \nlokal jika sedang mencari baju baru?")

    plt.show()

# Pilih variabel x dan y untuk regresi linear
x_var = 'Seberapa besar pengaruh merek fashion lokal dalam mempengaruhi keputusan anda untuk membeli produk mereka?'
y_var = 'Seberapa sering anda membeli produk fashion lokal jika sedang mencari baju baru?'

# Plot regresi linear untuk setiap variabel
plot_linear_regression(data[x_var], data[y_var], 'Regresi Linear: X horizontal dan Y vertikal', 'horizontal')
plot_linear_regression(data[x_var], data[y_var], 'Regresi Linear: X vertikal dan Y horizontal', 'vertical')
```
